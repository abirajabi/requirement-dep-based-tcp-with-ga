```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('itrust-500mid.csv', header=None)
df.columns = ["APFD", "DCR", "METHOD", "SUT", "TESTSET"]
print(df.head())
```

```python
sns.boxplot(data=df, x="SUT", y="APFD", hue="METHOD", palette="Set3")
plt.show()
```

```python
sns.boxplot(data=df, x="SUT", y="DCR", hue="METHOD", palette="Set3")
plt.show()
```
